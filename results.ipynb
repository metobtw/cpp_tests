```python
import pandas as pd

metaheuristics = ["SCA", "TLBO", "DE", "AOA", "ICA", "SSA", "WOA"]

data = {
    'Metaheuristic': metaheuristics,
    'PSNR Average': psnr_averages,
    'SSIM Average': ssim_averages,
    'Capacity Average': capacity_averages
}

df = pd.DataFrame(data)

print(df)
```

```python

# Сортируйте DataFrame по значению PSNR Average
df = df.sort_values(by='PSNR Average', ascending=False)

# Постройте график
fig, ax = plt.subplots()

# Example data
y_pos = range(len(df))

ax.barh(y_pos, df['PSNR Average'], align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(df['Metaheuristic'])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('PSNR')
ax.set_title('PSNR Comparison (Sorted)')

plt.show()
```
